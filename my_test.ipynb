```python
for i in tqdm(range(len(df))):
    url=df.loc[i,'url']
    if url_match_aff(url):
        df.at[i, 'isKnownLink'] = 1
    else:
        df.at[i, 'isKnownLink'] = 0
    if url_match_utm(url):
        df.at[i, 'isUTMlink'] = 1
    else:
        df.at[i, 'isUTMlink'] = 0
```
