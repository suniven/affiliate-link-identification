```python
import pandas as pd
import csv

csv.field_size_limit(500 * 1024 * 1024)
```

```python
df=pd.read_csv('youtube.csv', engine='python')
```

```python
now=df['url'].tolist()
```

```python
now
```

```python
df2=pd.read_csv('D:\\Python\\rebuild-db-for-youtube-url\\youtube_unique_filtered.csv', engine='python')
```

```python
all=df2['url'].tolist()
```

```python
from tqdm import tqdm
todo=[]
for i in tqdm(all):
    if i in now:
        continue
    else:
        todo.append(i)
```

```python
todo
```

```python
with open('./youtube_todo.txt','w',encoding='utf-8') as f:
    for i in todo:
        f.writelines(i+'\n')
```
