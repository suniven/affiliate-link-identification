```python
df_isAffiliate = df[~df['isKnownLink'].isna()]
df_maybeAffiliate = df[(~df['isUTMlink'].isna()) & df['isKnownLink'].isna()]
```

```python
affiliate_url_list = df_isAffiliate['url'].tolist()
affiliate_url_sus_list = df_maybeAffiliate['url'].tolist()
```

```python
twitter_all_urls=pd.read_csv('./twitter-all-urls.csv', engine='python')
```

```python
twitter_all_urls['isAffiliate']=0
```

```python
twitter_all_urls
```

```python
from tqdm import tqdm

for i in tqdm(range(len(twitter_all_urls))):
    if twitter_all_urls.loc[i, 'url'] in affiliate_url_list:
        twitter_all_urls.loc[i, 'isAffiliate']=1
    elif twitter_all_urls.loc[i, 'url'] in affiliate_url_sus_list:
        twitter_all_urls.loc[i, 'isAffiliate']=2
```

```python
twitter_all_urls
```

```python
twitter_all_urls[twitter_all_urls['isAffiliate']==1].shape
```

```python
twitter_all_urls[twitter_all_urls['isAffiliate']==2].shape
```

```python
df_twitter_all_affiliate_url = twitter_all_urls[twitter_all_urls['isAffiliate'] == 1]
df_twitter_all_sus_affiliate_url = twitter_all_urls[twitter_all_urls['isAffiliate'] == 2]
```

```python
df_twitter_all_affiliate_url.reset_index(drop=True)
df_twitter_all_sus_affiliate_url.reset_index(drop=True)
```

```python
df_twitter_all_affiliate_url['landing_page']=None
df_twitter_all_sus_affiliate_url['landing_page']=None
```

```python
for i in tqdm(range(len(df_twitter_all_affiliate_url))):
    current_url = df_twitter_all_affiliate_url.iloc[i, 0]
    current_landing_page = df[df['url'] == current_url].landing_page.tolist()
    if current_landing_page:
        df_twitter_all_affiliate_url.iloc[i, 2] = current_landing_page[0]
```

```python
for i in tqdm(range(len(df_twitter_all_sus_affiliate_url))):
    current_url = df_twitter_all_sus_affiliate_url.iloc[i, 0]
    current_landing_page = df[df['url'] == current_url].landing_page.tolist()
    if current_landing_page:
        df_twitter_all_sus_affiliate_url.iloc[i, 2] = current_landing_page[0]
```

```python
df_twitter_all_affiliate_url.reset_index(drop=True)
```
