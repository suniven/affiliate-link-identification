```python
import pandas as pd
import csv

csv.field_size_limit(500 * 1024 * 1024)

df = pd.read_csv('./youtube_all_results.csv', engine='python')
```

```python
df
```

```python
df=df[~df['isOffer'].isna()]
```

```python
df=df.reset_index(drop=True)
```

```python
df
```

```python
df_offer = pd.DataFrame(columns=['url', 'url_landing_page', 'landing_page_domain','offer_link','offer_landing_page'])
```

```python
df_offer
```

```python
df.loc[0, 'isOffer']
```

```python
from tqdm import tqdm
import ast
for i in tqdm(range(len(df))):
    url=df.loc[i, 'url']
    url_landing_page = df.loc[i, 'landing_page']
    landing_page_domain = df.loc[i,'landing_page_domain']
    offers = ast.literal_eval(df.loc[i, 'isOffer'])
    for offer_link, offer_landing_page in offers.items():
        item = {
            'url': url,
            'url_landing_page': url_landing_page,
            'landing_page_domain': landing_page_domain,
            'offer_link': offer_link,
            'offer_landing_page': offer_landing_page
        }
        df_offer.loc[len(df_offer)]=item
    # break   # for test
```
